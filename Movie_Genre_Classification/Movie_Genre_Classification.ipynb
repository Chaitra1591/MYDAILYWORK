```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("train_data.txt", sep=":::", engine="python",
                 names=["ID","TITLE","GENRE","PLOT"])

df.head()
```

```python
df.shape
```

```python
X = df["PLOT"]
y = df["GENRE"]
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)

X_tfidf = tfidf.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
def predict_genre(plot):
    vec = tfidf.transform([plot])
    pred = model.predict(vec)
    return str(pred[0]).strip()
```

```python
predict_genre("A young warrior fights dark forces to save his kingdom and protect his people")
```

```python
examples = [
    "A detective investigates a mysterious murder in a small town",
    "Two people fall in love despite coming from different worlds",
    "A group of astronauts travel through space to save humanity",
    "A warrior fights enemies to protect his kingdom"
]

for plot in examples:
    print("Plot:", plot)
    print("Predicted Genre:", predict_genre(plot))
    print()
```
